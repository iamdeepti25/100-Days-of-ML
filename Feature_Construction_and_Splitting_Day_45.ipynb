```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/titanic_train.csv',usecols=['Survived','Pclass','Age','SibSp','Parch'])
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
X = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
X,y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```

##Applying feature construction

```python
X['Family_size'] = X['Parch'] + X['SibSp'] + 1
```

```python

X.head()
```

```python
def myfun(num):
  if num == 1:
    return 0 #alone
  elif num> 1 and num <5:
    return 1 #small family
  else:
    return 2 #large family
```

```python
X['Family_type'] = X['Family_size'].apply(myfun)
```

```python
X.head()
```

```python
X.drop(columns=['SibSp','Parch','Family_size'],inplace=True)
```

```python
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```

###The accuracy increased after using feature construction

#Feature Splitting

```python
df = pd.read_csv('/content/titanic_train.csv')
```

```python
df.head()
```

```python
df['Name']
```

```python
df['Name']=df['Name'].str.split(',',expand=True)[1].str.split('.',expand=True)[1]
```

expandbool, default False
Expand the split strings into separate columns.

If True, return DataFrame/MultiIndex expanding dimensionality.

If False, return Series/Index, containing lists of strings.

```python
df['Title']=df['Name'].str.split(',',expand=True)[1].str.split('.',expand=True)[0]
```
