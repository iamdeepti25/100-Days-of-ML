When working with the `read_csv()` function in pandas, you may encounter situations where you need to specify the encoding parameter to ensure that the text in your CSV file is read correctly. Here are some guidelines to help you determine which encoding to use:

1. **UTF-8**: This is the most common encoding and supports a wide range of characters from different languages. If you're unsure about the encoding of your file, try UTF-8 first.

    ```python
    df = pd.read_csv('file.csv', encoding='utf-8')
    ```

2. **Latin-1 (ISO-8859-1)**: If you're working with Western European languages, Latin-1 might be a suitable choice.

    ```python
    df = pd.read_csv('file.csv', encoding='latin-1')
    ```

3. **UTF-16**: Use this encoding if you're dealing with languages that require two bytes to represent each character, such as Chinese, Japanese, or Korean.

    ```python
    df = pd.read_csv('file.csv', encoding='utf-16')
    ```

4. **Others**: If the above encodings don't work or you're working with a specific encoding, you may need to consult documentation or try different encodings until you find the correct one. Common encodings include UTF-16LE, UTF-16BE, UTF-32, Windows-1252, and more.

5. **Check the file**: Sometimes, the encoding of the file is explicitly stated in the file itself, especially for files generated by certain software. In such cases, you can open the file with a text editor and check for any encoding information at the beginning of the file.

6. **Try auto-detection**: You can also try using `encoding='latin1'` and let pandas attempt to detect the encoding automatically.

    ```python
    df = pd.read_csv('file.csv', encoding='latin1')
    ```

If you encounter decoding errors while reading the CSV file, it's a sign that the chosen encoding is incorrect, and you should try a different one until the file is read without errors.

When on_bad_lines is set to False, these "bad lines" are simply skipped, and the parsing process continues without raising an error.

##Converters

Functions for converting values in specified columns. Keys can either be column labels or column indices.

````python
def rename(name):
  if name == 'Royal Challengers Bangalore':
    return 'RCB'
  else:
     return name
````

##na_values parameter

If there are some place holders in a data set for example % or # and you want to conclude them as NaN or NA values, for that there is na_value parameter and there you can provide additional strings or place holders that should also be considered as NaN or NA values.

````python
pd.read_csv('aug_train.csv',na_values=['Male','Full time course','STEM'])
````
